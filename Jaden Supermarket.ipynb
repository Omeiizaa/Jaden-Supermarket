<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
    We have to add two useful columns to our dataset; Profit and Profit Margin

### **Steps:**
- Calculate **Profit** as the difference between Revenue and Cost.
- Derive **Profit Margin** to understand profitability relative to revenue.
</div>
</div>

```python
df['Profit'] = df['Revenue'] - df['Cost']
df['Profit Margin (%)'] = (df['Profit'] / df['Revenue']) * 100
```

```python
df.head()
```

<div style="text-align: justify; font-size: 20px; color: #007acc; font-family: Garamond; font-weight: bold">
   EXPLORATORY DATA ANALYSIS
</div>
<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
    We will explore the data to uncover patterns and trends in the dataset. We will look at Customer Demographics, Sales and Profit Trends, Regional Insights, and Customer segmentation

```python
#Distribution of numerical columns
print(df[['Customer Age', 'Cost', 'Revenue', 'Profit']].describe())

for col in ['Customer Age', 'Cost', 'Revenue', 'Profit']:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, color='#000080',)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#distribution of profit relationship with product category
plt.figure(figsize=(10,6))
sns.kdeplot(data=df, x="Profit", hue="Product Category", fill=True, alpha=0.5)
plt.title("Profit Distribution by Product Category", fontsize=16)
plt.xlabel("Profit", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# revenue and profit by product category
category_sales = df.groupby('Product Category')['Revenue'].sum().sort_values(ascending=False)
category_sales.plot(kind='bar', title='Revenue by Product Category', color='blue')
plt.ylabel('Total Revenue')
plt.show()

category_profits = df.groupby('Product Category')['Profit'].sum().sort_values(ascending=False)
category_profits.plot(kind='bar', title='Profit by Product Category', color='#000080')
plt.ylabel('Total Profit')
plt.show()
```

```python
# revenue and profit by product sub-category
subcategory_sales = df.groupby('Sub Category')['Revenue'].sum().sort_values(ascending=False)
subcategory_sales.plot(kind='bar', title='Revenue by Sub Category', color='blue')
plt.ylabel('Total Revenue')
plt.show()

subcategory_profits = df.groupby('Sub Category')['Profit'].sum().sort_values(ascending=False)
subcategory_profits.plot(kind='bar', title='Profit by Sub Category', color='#000080')
plt.ylabel('Total Profit')
plt.show()
```

```python
#top country and state by profit

country_profit = df.groupby('Country')['Profit'].sum().sort_values(ascending=False)
country_profit.head(10).plot(kind='bar', title='Top Countries by Profit', color='blue')
plt.ylabel('Total Profit')
plt.show()

state_profit = df.groupby('State')['Profit'].sum().sort_values(ascending=False)
state_profit.head(10).plot(kind='bar', title='Top States by Profit', color='#000080')
plt.ylabel('Total Profit')
plt.show()
```

```python
#gender distribution
gender_counts = df['Customer Gender'].value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', title='Customer Gender Distribution', colors=['blue', '#007FFF'])
plt.ylabel('')
plt.show()
```
